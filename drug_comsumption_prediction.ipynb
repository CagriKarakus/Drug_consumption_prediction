```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# scikit-learn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Model
from sklearn.linear_model import RidgeClassifier

# BayesÃ§i hiperparametre aramasÄ±
from skopt import BayesSearchCV

# -------------------------------------------------------
# 1) Veri YÃ¼kleme ve Kolon Ä°simlerinin Belirlenmesi
# -------------------------------------------------------
column_names = [
    "ID", "Age", "Gender", "Education", "Country", "Ethnicity",
    "Nscore", "Escore", "Oscore", "Ascore", "Cscore",
    "Impulsive", "SS",
    "Alcohol", "Amphet", "Amyl", "Benzos", "Caff",
    "Cannabis", "Choc", "Coke", "Crack", "Ecstasy",
    "Heroin", "Ketamine", "Legalh", "LSD", "Meth",
    "Mushrooms", "Nicotine", "Semer", "VSA"
]

data = pd.read_csv('/content/drug_consumption.data', header=None, names=column_names)

# -------------------------------------------------------
# 2) Ã–zelliklerin (Features) SeÃ§ilmesi ve Kodlama/Ã–lÃ§eklendirme
# -------------------------------------------------------
X = data[[
    "Age", "Gender", "Education", "Country", "Ethnicity",
    "Nscore", "Escore", "Oscore", "Ascore", "Cscore",
    "Impulsive", "SS"
]]
substance_columns = column_names[13:]  # 13. kolondan sonraki tÃ¼m sÃ¼tunlar

# KullanÄ±m seviyelerini 3 sÄ±nÄ±fa indirgeme (Ã¶rnek kodlama)
for col in substance_columns:
    data[col] = data[col].replace({
        'CL0': 0, 'CL1': 1,           # DÃ¼ÅŸÃ¼k kullanÄ±m
        'CL2': 2, 'CL3': 3, 'CL4': 4, # Orta kullanÄ±m
        'CL5': 5, 'CL6': 6            # YÃ¼ksek kullanÄ±m
    })

# Ã–zellikleri Ã¶lÃ§eklendirme
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# -------------------------------------------------------
# 3) BayesÃ§i Arama Ä°Ã§in Hiperparametre AlanÄ±
# -------------------------------------------------------
# RidgeClassifier -> alpha parametresini log-uniform aralÄ±kta arayacaÄŸÄ±z (1e-5 ~ 1e3)
search_spaces = {
    'alpha': (1e-5, 1e3, 'log-uniform')
}

# -------------------------------------------------------
# 4) Her Madde Ä°Ã§in RidgeClassifier + Bayesian Optimization
# -------------------------------------------------------
for substance in substance_columns:
    print(f"\n===== Substance: {substance} =====")

    y = data[substance]

    # Kategorik deÄŸerleri sayÄ±sala Ã§evir
    y_encoded, y_classes = pd.factorize(y)

    # EÄŸitim ve test setlerine ayÄ±rma
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y_encoded,
        test_size=0.3,
        random_state=42
    )

    # BayesSearchCV ile RidgeClassifier
    # n_iter: kaÃ§ farklÄ± hiperparametre deneneceÄŸi
    bayes_search = BayesSearchCV(
        estimator=RidgeClassifier(),
        search_spaces=search_spaces,
        n_iter=32,               # Daha fazla arttÄ±rÄ±rsanÄ±z daha iyi sonuÃ§lar bulabilir (zamana baÄŸlÄ±)
        cv=5,                    # 5 katlÄ± Ã§apraz doÄŸrulama
        scoring='accuracy',
        n_jobs=-1,               # TÃ¼m CPU Ã§ekirdeklerini kullan
        random_state=42
    )

    bayes_search.fit(X_train, y_train)
    best_model = bayes_search.best_estimator_
    best_alpha = bayes_search.best_params_['alpha']

    print(f"Best alpha (BayesSearch): {best_alpha:.4f}")

    # Modeli test setinde deÄŸerlendirme
    y_pred = best_model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy (DoÄŸruluk): {accuracy:.3f}")

    # YalnÄ±zca test setinde var olan sÄ±nÄ±flarÄ± bulalÄ±m
    unique_label_indices = np.unique(y_test)

    # Classification Report
    report = classification_report(
        y_test,
        y_pred,
        labels=unique_label_indices,
        target_names=[str(y_classes[i]) for i in unique_label_indices],
        zero_division=0
    )
    print("Classification Report:")
    print(report)

    # ----------------------------------------------------
    # 4.1) Confusion Matrix (KarmaÅŸÄ±klÄ±k Matrisi) GrafiÄŸi
    # ----------------------------------------------------
    cm = confusion_matrix(y_test, y_pred, labels=unique_label_indices)
    plt.figure(figsize=(7, 5))
    sns.heatmap(
        cm,
        annot=True,
        fmt="d",
        cmap="Blues",
        xticklabels=[str(y_classes[i]) for i in unique_label_indices],
        yticklabels=[str(y_classes[i]) for i in unique_label_indices]
    )
    plt.title(f"Confusion Matrix - {substance}")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.tight_layout()
    plt.show()

    # ----------------------------------------------------
    # 4.2) Ã–zellik KatsayÄ±larÄ± Bar GrafiÄŸi
    # ----------------------------------------------------
    # RidgeClassifier'Ä±n katmanlarÄ± (coef_) bir matriste saklanÄ±r
    # Ä°lk satÄ±r, "0" sÄ±nÄ±fÄ± ile ilgili katsayÄ±larÄ± temsil eder (binary senaryoda)
    # 3 sÄ±nÄ±fÄ±nÄ±z varsa shape (3, n_features) olabilir.
    # Bu kod, illÃ¼stratif olarak ilk satÄ±rÄ± (0. index) gÃ¶steriyor.
    coefficients = best_model.coef_[0]

    plt.figure(figsize=(12, 5))
    plt.bar(X.columns, coefficients, color='skyblue', edgecolor='black')
    plt.title(f'Feature Influence on {substance} (Accuracy: {accuracy:.3f}, alpha={best_alpha:.4f})', fontsize=14)
    plt.xlabel('Features', fontsize=12)
    plt.ylabel('Coefficient Value', fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

    # ----------------------------------------------------
    # 4.3) Classification Report'un Bar Grafikleri
    # ----------------------------------------------------
    # Raporu sÃ¶zlÃ¼k formatÄ±na Ã§evirip DataFrame'e dÃ¶nÃ¼ÅŸtÃ¼relim
    report_dict = classification_report(
        y_test,
        y_pred,
        labels=unique_label_indices,
        target_names=[str(y_classes[i]) for i in unique_label_indices],
        zero_division=0,
        output_dict=True
    )
    df_report = pd.DataFrame(report_dict).transpose()

    # "accuracy", "macro avg", "weighted avg" vb. hariÃ§ tutarak, sadece sÄ±nÄ±flarÄ± alacaÄŸÄ±z
    df_class_report = df_report.drop(['accuracy', 'macro avg', 'weighted avg'], errors='ignore')

    metrics_to_plot = ['precision', 'recall', 'f1-score']
    plt.figure(figsize=(8, 5))
    df_class_report[metrics_to_plot].plot(kind='bar')
    plt.title(f'{substance} - Precision, Recall, F1-Score by Class', fontsize=14)
    plt.xlabel('Classes', fontsize=12)
    plt.ylabel('Score', fontsize=12)
    plt.xticks(rotation=0)
    plt.ylim([0, 1])
    plt.legend(loc='lower right')
    plt.tight_layout()
    plt.show()

    # Support grafiÄŸi
    plt.figure(figsize=(6, 4))
    df_class_report[['support']].plot(kind='bar', legend=False)
    plt.title(f'{substance} - Support (Num of Samples by Class)', fontsize=14)
    plt.xlabel('Classes', fontsize=12)
    plt.ylabel('Support Count', fontsize=12)
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()

    print("\n")  # Konsolda ayrÄ±m iÃ§in boÅŸluk
```
